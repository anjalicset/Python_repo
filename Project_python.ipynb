```python
# Scraping from Glassdoor
job_title=['job_title']
company=['company']
location=['location']
rating=['rating']
salary=['salary']
final_list = []
jobs=["data-scientist", "data-analyst"]
for job in jobs:
    number = 0
    if job == 'data-scientist':
        number = 14
    else:
        number = 12
        
    for i in range(1, 26):
        source = requests.get(
            'https://www.glassdoor.ca/Job/' + str(job) + '-jobs-SRCH_KO0,' + 
            str(number) + '_IP' + str(i) + '.htm').text
        
        soup = BeautifulSoup(source, 'lxml')
        
        job_containers = soup.find_all('li', class_ = 'react-job-listing')

        for container in job_containers:
            title_t = container.find('a', class_='jobLink css-1rd3saf eigr9kq3').text

            if container.find('a', class_='css-l2wjgv e1n63ojh0 jobLink'):
                company_t = container.find('a', class_='css-l2wjgv e1n63ojh0 jobLink').text
            else:
                company_t = 'NA'

            if container.find('span', class_='css-3g3psg pr-xxsm css-iii9i8 e1rrn5ka0'):
                location_t = container.find('span', class_='css-3g3psg pr-xxsm css-iii9i8 e1rrn5ka0').text
            else:
                location_t = 'NA'

            if container.find('span', class_='css-2lqh28 e1cjmv6j1'):
                rating_t = container.find('span', class_='css-2lqh28 e1cjmv6j1').text
            else:
                rating_t = 'NA'

            if container.find('span', class_='css-1xe2xww e1wijj242'):
                salary_t = container.find('span', class_='css-1xe2xww e1wijj242').text
                salary_t = salary_t.split('(')
                salary_t = salary_t[0].strip()
            else:
                salary_t = 'NA'
            row_data = [title_t,company_t,location_t,rating_t,salary_t]
            final_list = final_list + row_data

final_list
```
